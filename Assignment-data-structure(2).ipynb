````python
#5Describe the key features of sets and provide examples of their use
Definition: A set is a collection of distinct objects, considered as an object in its own right. The objects in a set are called elements or members.

#Uniqueness: Each element in a set must be unique. No duplicates are allowed.
#Order: The order of elements in a set does not matter. {1, 2, 3} is the same set as {3, 2, 1}.
#Notation: Sets are usually denoted by curly brackets. For example, {a, b, c} is a set with three elements: a, b, and c.
#Empty Set: The empty set, denoted by {} or ∅, is a set with no elements.
#Membership: If an element a is in set S, we write a ∈ S. If a is not in S, we write a ∉ S.
#Subset: A set A is a subset of set B if all elements of A are also elements of B. This is denoted as A ⊆ B.
#Union: The union of two sets A and B, denoted A ∪ B, is a set containing all elements from both A and B.
#Intersection: The intersection of two sets A and B, denoted A ∩ B, is a set containing only the elements that are in both A and B.
#Difference: The difference between two sets A and B, denoted A - B, is a set containing elements that are in A but not in B.
#Complement: The complement of a set A refers to elements not in A, relative to a universal set U that contains all possible elements of interest

#Adding and Removing Elements:

set1 = {1, 2, 3}
set1.add(4)  # Adds 4 to the set
print(set1)  # Output: {1, 2, 3, 4}
set1.discard(2)  # Removes 2 from the set, no error if 2 is not present
print(set1)  # Output: {1, 3, 4}
set1.remove(1)  # Removes 1 from the set, raises KeyError if 1 is not present
print(set1)  # Output: {3, 4}

#Unordered:
fruits = {"apple", "banana", "cherry"}
print(fruits)  # Output: {'cherry', 'banana', 'apple'} (order can vary)

#unique elements
numbers = {1, 2, 2, 3, 4, 4, 5}
print(numbers)  # Output: {1, 2, 3, 4, 5}
 
 #mutable
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)  # Output: {'cherry', 'banana', 'apple'}

fruits.remove("banana")
print(fruits)  # Output: {'cherry', 'apple'}
````

````python
# 7.Describe how to add, modify, and delete items in a dictionary with examples


# Adding Items

You can add items to a dictionary by assigning a value to a new key. If the key already exists, the value will be updated.

**Example**:
```python
# Create an empty dictionary
person = {}

# Add new key-value pairs
person["name"] = "Alice"
person["age"] = 30
person["city"] = "New York"

print(person)
# Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}


# Modifying Items

You modify items in a dictionary by reassigning a new value to an existing key.

**Example**:
python
# Create a dictionary with some initial items
person = {"name": "Alice", "age": 30, "city": "New York"}

# Modify the value associated with an existing key
person["age"] = 31
person["city"] = "San Francisco"

print(person)
# Output: {'name': 'Alice', 'age': 31, 'city': 'San Francisco'}


# Deleting Items

You can delete items from a dictionary using various methods, such as the `del` statement, the `pop()` method, or the `popitem()` method.

**1. Using `del`**:
The `del` statement removes a specific key-value pair from the dictionary.

**Example**:
```python
# Create a dictionary with some items
person = {"name": "Alice", "age": 30, "city": "New York"}

# Delete a key-value pair
del person["city"]

print(person)
# Output: {'name': 'Alice', 'age': 30}
```

**2. Using `pop()`**:
The `pop()` method removes a specific key and returns its value. If the key does not exist, it raises a `KeyError` unless a default value is provided.

**Example**:
```python
# Create a dictionary with some items
person = {"name": "Alice", "age": 30, "city": "New York"}

# Remove a key-value pair and get its value
age = person.pop("age")

print(person)
# Output: {'name': 'Alice', 'city': 'New York'}
print("Removed age:", age)
# Output: Removed age: 30
```

**3. Using `popitem()`**:
The `popitem()` method removes and returns the last key-value pair added to the dictionary. It raises a `KeyError` if the dictionary is empty.

**Example**:
```python
# Create a dictionary with some items
person = {"name": "Alice", "age": 30, "city": "New York"}

# Remove and return the last key-value pair
last_item = person.popitem()

print(person)
# Output: {'name': 'Alice', 'age': 30}
print("Removed item:", last_item)
# Output: Removed item: ('city', 'New York')
```

 
````

````python
#8.Discuss the importance of dictionary keys being immutable and provide examples.
#ans Importance of Immutable Keys

    Hashing:
        Explanation: Dictionaries in Python are implemented using hash tables. This means that each key is hashed to determine its position in the table. The immutability of dictionary keys ensures that their hash value remains constant during their lifetime, which is essential for consistent and reliable key-based access.
        Example: If a key were mutable and its hash value changed, it could cause inconsistencies in data retrieval. For instance:

        python

    d = {}
    key = [1, 2, 3]  # Lists are mutable and thus cannot be dictionary keys
    d[key] = "value"  # This will raise a TypeError

Consistency:

    Explanation: Immutability ensures that once a key is added to the dictionary, its value and position in the hash table remain consistent. If keys were mutable, changes to the key could lead to errors or difficulties in locating the key-value pair.
    Example: Consider a dictionary where keys are mutable and can change:

    python

    d = {}
    key = (1, 2, 3)  # Tuples are immutable and can be used as dictionary keys
    d[key] = "value"

   
 Hash Table Integrity:

    Explanation: Hash tables use keys to determine the index where values are stored. If keys were mutable, the hash values could change, potentially leading to corruption or loss of data in the hash table.
    Example: Using mutable types such as lists as keys would not work because lists can change in size or content, which would affect their hash values and break the dictionary’s indexing:

    python

        d = {}
        key = [1, 2, 3]
        try:
            d[key] = "value"
        except TypeError as e:
            print(e)  # Output: unhashable type: 'list'

Examples of Immutable and Mutable Keys

Immutable Keys:

    Integers:

    python

d = {}
d[1] = "one"
print(d)  # Output: {1: 'one'}

Strings:

python

d = {}
d["name"] = "Alice"
print(d)  # Output: {'name': 'Alice'}

Tuples:

python

    d = {}
    d[(1, 2, 3)] = "tuple_key"
    print(d)  # Output: {(1, 2, 3): 'tuple_key'}

Mutable Keys (Not Allowed):

    Lists:

    python

d = {}
try:
    d[[1, 2, 3]] = "list_key"
except TypeError as e:
    print(e)  # Output: unhashable type: 'list'

Dictionaries:

python

d = {}
try:
    d[{"key": "value"}] = "dict_key"
except TypeError as e:
    print(e)  # Output: unhashable type: 'dict'
````
